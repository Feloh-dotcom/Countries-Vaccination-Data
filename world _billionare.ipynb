DATA CLEANING

```python
cols = list(Data.columns)
```

```python
birth_cols = ['birthYear', 'birthMonth', 'birthDay']
```

```python
birth_positions = [cols.index(col) for col in birth_cols]
```

```python
new_birth_order = ['birthDay', 'birthMonth', 'birthYear']
```

```python
for pos, col in zip(sorted(birth_positions), new_birth_order):
    cols[pos] = col
```

```python
Data = Data[cols]
```

```python
Data.head()
```

```python
Data['Date of birth'] = Data['birthDay'].astype(str) + '/' + \
                        Data['birthMonth'].astype(str) + '/' + \
                        Data['birthYear'].astype(str)
```

```python
Data.head()
```

```python
Data = Data.drop(columns=['birthDay', 'birthMonth', 'birthYear'])
```

```python
Data.head()
```

```python
Data = Data.drop(columns=['lastName', 'firstName'])
```

```python
Data = Data.rename(columns={'personName': 'Full_Name'})
```

```python
Data['gender'] = Data['gender'].replace({'M': 'Male', 'F': 'Female'})
```

```python
Data['Age']=0
```

```python
from datetime import datetime
```

```python
Data['Date'] = datetime.now().strftime('%d/%m/%y')
```

```python
Data = Data.drop(columns=['Age'])
```

```python
# Convert 'Date of birth' to datetime
Data['Date of birth'] = pd.to_datetime(Data['Date of birth'], format='%d/%m/%Y', errors='coerce')

# Calculate age
today = datetime.now()
Data['age'] = Data['Date of birth'].apply(
    lambda dob: today.year - dob.year - ((today.month, today.day) < (dob.month, dob.day)) 
    if pd.notnull(dob) else None
)

# Display full dataset (with all columns, including 'age')
```

```python
Data.head()
```

```python
# Remove ',' and '$', then convert to integers
Data['gdp_country'] = (
    Data['gdp_country']
    .astype(str)                # ensure values are strings
    .str.replace(',', '', regex=False)  # remove commas
    .str.replace('$', '', regex=False)  # remove dollar signs
    .astype(float)              # convert to float first (to handle decimals)
    .astype(int)                # finally convert to integer
)
```

```python
Data.head()
```

```python
Data = Data.rename(columns={'finalWorth': 'Net Worth'})
```

```python
import matplotlib.pyplot as plt
```

```python
top10 = Data[['age', 'NetWorth']].sort_values(by='NetWorth', ascending=False).head(10)
```

```python
# Create the bar chart
plt.figure(figsize=(10,6))
plt.bar(top10['age'].astype(str), top10['NetWorth'])  # convert age to string for x-axis labels
plt.xlabel('Age')
plt.ylabel('NetWorth')
plt.title('Top 10 Richest People: Age vs NetWorth')
plt.xticks(rotation=45)  # rotate x labels for readability
plt.tight_layout()
plt.show()
```

```python
# Aggregate netWorth by source and get top 10
source_networth = Data.groupby('source')['NetWorth'].sum().sort_values(ascending=False).head(10)

# Plot the bar chart
plt.figure(figsize=(12,6))
plt.bar(source_networth.index, source_networth.values, color='skyblue')
plt.xlabel('Source')
plt.ylabel('(Net Worth in $)')
plt.title('Top 10 Richest Sources by Net Worth')
plt.xticks(rotation=45, ha='right')  # rotate x-axis labels for readability
plt.tight_layout()
plt.show()
```

```python
gender_networth = Data.groupby('gender')['NetWorth'].sum()

# Plot the bar chart
plt.figure(figsize=(8,6))
plt.bar(gender_networth.index, gender_networth.values, color=['skyblue', 'lightpink'])
plt.xlabel('Gender')
plt.ylabel('Net Worth')
plt.title('Net Worth by Gender')
plt.tight_layout()
plt.show()
```
